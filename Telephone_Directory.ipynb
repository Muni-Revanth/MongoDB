# Telephone Directory 

## Importing Necessary Libraries

```python
import pymongo
import pandas as pd
```

## Activating MongoClient 

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
```

## Creating a new Database

```python
tel_dir = client.Telephone_Directory
```

## Creating a new Collection

```python
tel_info = tel_dir.supermart_info
```

## Importing Data

```python
df = pd.read_json("R:\\Supermart.json")
```

```python
df_dict = df.to_dict("records")
```

```python
tel_info.insert_many(df_dict)
```

## Creating records 

```python
Customer_info = [{"name":"raghu", "mobile":"90*******4","place":"chennai","state":"tamilnadu"},{"name":"ramesh", "mobile":"77*******6","place":"tanjore","state":"tamilnadu"},{"name":"gowri", "mobile":"89*******1","place":"mysore","state":"karnataka"},{"name":"farook", "mobile":"91*******9","place":"ongole","state":"karnataka"}]
```

## Creating a new Collection

```python
tel_customers = tel_dir.Customer_Contacts
```

```python
tel_customers.insert_many(Customer_info)
```

## Inserting a new record into a collection 

```python
add_cust = {"name":"stephen", "mobile":"63*******2","place":"hyderabad","state":"telangana"}
```

```python
tel_customers.insert_one(add_cust)
```

## Retrieving Operation 

```python
t = tel_customers.find({"name":"farook"})
for y in t:
    print(y)
```

## Updating Operation 

```python
u = tel_info.find({"Customer Name":"Harish"})
for i in u:
    print(i)
```

```python
tel_info.update_one({"Customer Name":"Harish"}, {"$inc":{"Sales":6}}, upsert=True)
```

```python
u = tel_info.find({"Customer Name":"Harish"})
for i in u:
    print(i)
```

## Delete Operation 

```python
d = tel_info.find({"Customer Name":"Sudha"})
for j in d:
    print(j)
```

```python
tel_info.delete_one({"Customer Name":"Sudha"})
```

```python
d = tel_info.find({"Customer Name":"Sudha"})
for j in d:
    print(j)
```
